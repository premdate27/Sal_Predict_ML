```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in ['Gender', 'Education Level', 'Job Title']:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
x = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8,test_size=0.2, random_state=2529)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score

# 1. Linear Regression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
score_lr = r2_score(y_test, y_pred_lr)

# 2. Decision Tree Regressor
dt = DecisionTreeRegressor(random_state=2529)
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)
score_dt = r2_score(y_test, y_pred_dt)

# 3. Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=2529)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
score_rf = r2_score(y_test, y_pred_rf)

# 4. Gradient Boosting Regressor
gb = GradientBoostingRegressor(random_state=2529)
gb.fit(x_train, y_train)
y_pred_gb = gb.predict(x_test)
score_gb = r2_score(y_test, y_pred_gb)

# Display Results
print(f"Linear Regression R2 Score:    {score_lr:.4f}")
print(f"Decision Tree R2 Score:      {score_dt:.4f}")
print(f"Random Forest R2 Score:      {score_rf:.4f}")
print(f"Gradient Boosting R2 Score:  {score_gb:.4f}")
```

```python
import pickle

# Save the best model (Random Forest) to a file
with open('best_salary_model.pkl', 'wb') as f:
    pickle.dump(rf, f)

print("Best model (Random Forest) saved successfully as 'best_salary_model.pkl'")
```
